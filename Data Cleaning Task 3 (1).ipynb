## Clean the data..(missing values,outliers,type casting,duplicate,variance,...etc)
## missing values

```python
df.isna().sum()
```

```python
max_id = df['Customer ID'].max()
df['Customer ID'].fillna(max_id + 1,inplace = True)
```

```python
df['Age'].fillna(df['Age'].mode()[0],inplace = True)  #Age

df['Gender'].fillna(df['Gender'].mode()[0],inplace = True)  #Gender

df['Item Purchased'].fillna(df['Item Purchased'].mode()[0],inplace = True)  #Item Purchased

df["Category"].fillna(df["Category"].mode()[0],inplace = True) # category

df["Location"].fillna(df["Location"].mode()[0],inplace = True) # location

df["Color"].fillna(df["Color"].mode()[0],inplace = True) # color

df["Season"].fillna(df["Season"].mode()[0],inplace = True)  # season

df["Review Rating"].fillna(df["Review Rating"].mode()[0],inplace = True) # review rating

df["Subscription Status"].fillna(df["Subscription Status"].mode()[0],inplace = True) # subscription status

df["Discount Applied"].fillna(df["Discount Applied"].mode()[0],inplace = True) # discount applied

df["Promo Code Used"].fillna(df["Promo Code Used"].mode()[0],inplace = True)  # promo code used

df["Payment Method"].fillna(df["Payment Method"].mode()[0],inplace = True) # payment metod

df["Frequency of Purchases"].fillna(df["Frequency of Purchases"].mode()[0],inplace = True) # Frequency of Purchases
```

## Outliers

```python
df[(df["Age"] < 0) | (df["Age"] > 100)] # age

df["Age"].fillna(df["Age"].mode()[0],inplace = True)

condition = (df["Review Rating"] < 0) | (df["Review Rating"] > 5)   # review rating

df["Review Rating"] = np.where(condition,3.5,df["Review Rating"])
```

## Scaling

```python
df["Age"].mean(),df["Age"].median()  # age
```

```python
def standard(x):    # age
    n = (x - df["Age"].mean()) / df["Age"].std()
    return n
df["Age"] = df["Age"].apply(standard)
```

```python
df["Purchase Amount (USD)"].mean(),df["Purchase Amount (USD)"].median()  # Purchase Amount (USD)
```

```python
def standard(x):    # Purchase Amount (USD)
    n = (x - df["Purchase Amount (USD)"].mean()) / df["Purchase Amount (USD)"].std()
    return n
df["Purchase Amount (USD)"] = df["Purchase Amount (USD)"].apply(standard)
```

```python
df["Review Rating"].mean(),df["Review Rating"].median()  # review rating
```

```python
def standard(x):    # review rating
    n = (x - df["Review Rating"].mean()) / df["Review Rating"].std()
    return n
df["Review Rating"] = df["Review Rating"].apply(standard)
```
